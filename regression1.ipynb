```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Show the first five rows of the data set.
# Show the description and the info of the data set.
temp_data = pd.read_csv('Height_Weight_single_variable_data_101_series_1.0.csv')
temp_data.head()
```

```python
X = temp_data['Weight'].values
y = temp_data['Height'].values
```

```python
X = X.reshape(-1, 1)
X
```

```python
# Using a regression model, split your data into train and test data. 
# split (80/20)one for testing and one for training to make presumptions, and compare
x_train, x_test, y_train, y_test = train_test_split(X, y, train_size=.8, test_size=.2, random_state=100)
```

```python
print(f'X_Train shape [{x_train.shape}]')
print(f'X_Test shape [{x_test.shape}]')
print(f'y_Train shape [{y_train.shape}]')
print(f'y_Test shape [{y_test.shape}]')
```

```python
# Fit your training split to the regression model.

plt.scatter(x_train, y_train, color='blue')
plt.xlabel('Weight')
plt.ylabel('Height')
plt.title('Height Vs. Weight')
```

```python
#  linear regression
lm = LinearRegression()
lm.fit(x_train, y_train)
```

```python
y_predict = lm.predict(x_test)
```

```python
# Show your regression model’s score.
print(f'Height vs Weight Train Accuracy {round(lm.score(x_train, y_train)* 100,2)}%')
print(f'Height vs Weight Test Accuracy {round(lm.score(x_test, y_test)* 100,2)}%')
```

```python
plt.scatter(x_train, y_train, color='teal')
plt.plot(x_test, y_predict, color='tomato')
plt.xlabel('Weight')
plt.ylabel('Height')
plt.title('Height Vs. Weight')
```
