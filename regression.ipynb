```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
temp_data = pd.read_csv('Environment_Temperature_change_E_All_Data_NOFLAG.csv')
temp_data.head()
```

```python
# selecting features
X = temp_data['Y1961'].values
y = temp_data['Months'].values
```

```python
# X data
X
```

```python
# y data
y
```

```python
X = X.reshape(-1, 1)
X
```

```python
#  X feature = 1961
#  y feature = 2019
#  split (80/20)one for testing and one for training to make presumptions, and compare
x_train, x_test, y_train, y_test = train_test_split(X, y, train_size=.8, test_size=.2, random_state=100)
```

```python
print(f'X_Train shape [{x_train.shape}]')
print(f'X_Test shape [{x_test.shape}]')
print(f'y_Train shape [{y_train.shape}]')
print(f'y_Test shape [{y_test.shape}]')
```

```python
plt.scatter(x_train, y_train, color='blue')
plt.xlabel('Temperature Deviations')
plt.ylabel('Months in 1961')
```
